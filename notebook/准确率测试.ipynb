```python
import pandas as pd
test_data=pd.read_csv(r'../data/test300_result_verify.csv')
# display(test_data)
jieba_true=0
ltp_true=0
false_index=[]
for index,row in test_data.iterrows():
    if row['flag']==row['jieba']:
        jieba_true+=1
    if row['flag']==row['ltp']:
        ltp_true+=1
    if row['flag']!=row['jieba'] and row['flag']!=row['ltp']:
        false_index.append(index)
print('jieba_accuracy:',jieba_true/len(test_data))
print('ltp_accuracy:',ltp_true/len(test_data))
print('both_false:',len(false_index)/len(test_data))
# for i in false_index:
#     print (i)
```

```python
import pandas as pd
test_data=pd.read_csv(r'../data/baidu_new_test300_result_verify.csv')
# display(test_data)
jieba_true=0
ltp_true=0
baidu_true=0
false_index=[]
baidu_none=[]
for index,row in test_data.iterrows():
    if row['flag']==row['jieba']:
        jieba_true+=1
    if row['flag']==row['ltp']:
        ltp_true+=1
    if row['flag']==row['baidu']:
        baidu_true+=1 
    if row['baidu']==8:
        baidu_none.append(index)   
    if row['flag']!=row['jieba'] and row['flag']!=row['ltp']:
        false_index.append(index)
#         print(index,row['danmu'])
print('jieba_accuracy:',jieba_true/len(test_data))
print('ltp_accuracy:',ltp_true/len(test_data))
print('both_false:',len(false_index)/len(test_data))
print('baidu_none:',len(baidu_none))
print('baidu_accuracy:',baidu_true/(len(test_data)-len(baidu_none)))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
flag=test_data['flag'].tolist()
jieba=test_data['jieba'].tolist()
ltp=test_data['ltp'].tolist()
target=['-1','0','1']
print('jieba',accuracy_score(flag,jieba))
print('ltp',accuracy_score(flag,ltp))
print('jieba',classification_report(flag,jieba,target_names=target))
print('ltp',classification_report(flag,ltp,target_names=target))
print('jieba',confusion_matrix(flag,jieba))
print('ltp',confusion_matrix(flag,ltp))
```
