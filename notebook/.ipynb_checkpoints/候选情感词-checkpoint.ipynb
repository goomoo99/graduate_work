```python
import pandas as pd
jieba=pd.read_csv(r'../data/freq_st_jieba_cleaned_room36252danmu_500000.csv',header=None)
# display(jieba)
ltp=pd.read_csv(r'../data/freq_st_ltp_cleaned_room36252danmu_500000.csv',header=None)
# display(ltp)
# Âà†ÂéªÈ¢ëÁéáÂ∞è‰∫é10ÁöÑËØç
jieba_dele=[]
ltp_dele=[]

for index,row in jieba.iterrows():
    if row[1]<10:
        jieba_dele.append(index)

for index,row in ltp.iterrows():
    if row[1]<10:
        ltp_dele.append(index)
print(len(jieba_dele),len(ltp_dele)) 

jieba=jieba.drop(index=jieba_dele)
ltp=ltp.drop(index=ltp_dele)
# display(jieba)
# display(ltp)
```

```python
# ÂêàÂπ∂„ÄÅÂéªÈáç
candidate=jieba.append(ltp)

candidate.columns=['word','freq']
candidate.sort_values(by='freq',ascending=False,inplace=True)

candidate.drop_duplicates(subset=['word'],keep='first',inplace=True)
display(candidate)
candidate.to_csv(r'../data/candidate_sentiment.csv',header=None,index=None)
```

```python
# ÊñóÈ±ºÊÉÖÊÑüËØçÂä†ÂÖ•Ëá™ÂÆö‰πâËØçÂÖ∏
self_word=pd.read_csv(r'../dict/self_dict.txt',header=None)
# display(game_word)
self_word=self_word.drop_duplicates()
display(self_word)
self_word.to_csv(r'../dict/self_dict.txt',header=None,index=None)
```

```python
# ÂÄôÈÄâËØçÂéªÈô§Ê∏∏ÊàèÊúØËØ≠ËØç
candidat=pd.read_csv(r'../data/candidate_sentiment.csv',header=None)

game_word=pd.read_csv(r'../dict/game_word.txt',header=None)
# display(game_word)
game_word=game_word[0].tolist()
len(game_word)
dele=[]
for index,row in candidat.iterrows():
    if row[0] in game_word:
        dele.append(index)
candidat.drop(index=dele,inplace=True)
display(candidat)
# print(len(dele))        
```

```python
# ÂÄôÈÄâËØçÂéªÈô§Â∑≤Áü•ÊÉÖÊÑüËØç
pos=pd.read_csv(r'../dict/self_positive_dict.txt',header=None)
neg=pd.read_csv(r'../dict/self_negative_dict.txt',header=None)
# display(pos)
# display(neg)
pos_list=pos[0].tolist()
neg_list=neg[0].tolist()
print(len(pos_list),len(neg_list))
dele=[]
# display(candidat)
for index,row in candidat.iterrows():
    if row[0] in pos_list or row[0] in neg_list:
        dele.append(index)

print(len(dele))        
candidat=candidat.drop(index=dele)
display(candidat)
candidat.to_csv(r'../data/candidate_sentiment.csv',header=None,index=None)
```

```python
# ËØçÊÄßÊ†áÊ≥®-ltp
import os
candidate=pd.read_csv(r'../data/candidate_sentiment.csv',header=None)
can_word=candidate[0].tolist()
# Êñ∞Âä†‰∏ÄÂàóÂ≠òÊîæËØçÊÄß
candidate.insert(2,'ltp_pos',0)
# display(candidate)
LTP_DATA_DIR = '../ltp_data_v3.4.0/ltp_data_v3.4.0'  # ltpÊ®°ÂûãÁõÆÂΩïÁöÑË∑ØÂæÑ
pos_model_path = os.path.join(LTP_DATA_DIR, 'pos.model')  # ËØçÊÄßÊ†áÊ≥®Ê®°ÂûãË∑ØÂæÑÔºåÊ®°ÂûãÂêçÁß∞‰∏∫`pos.model`

from pyltp import Postagger
postagger = Postagger() # ÂàùÂßãÂåñÂÆû‰æã
postagger.load(pos_model_path)  # Âä†ËΩΩÊ®°Âûã


postags = postagger.postag(can_word)  # ËØçÊÄßÊ†áÊ≥®

postagger.release()  # ÈáäÊîæÊ®°Âûã

postags=list(postags)
candidate['ltp_pos']=postags
display(candidate)
```

```python
# ËØçÊÄßÊ†áÊ≥®-jieba

# candidate.insert(3,'jieba_pos',0)
# candidate.columns=['word','freq','ltp_pos','jieba_pos']
# display(candidate)


import jieba.posseg as pseg
# words =pseg.cut("‰∏ç")
# for w in words:
#     print (w.flag)
jieba_pos=[]
for index,row in candidate.iterrows():
    s=row['word']
    words=pseg.cut(s)
    pos=[]
    for w in words:
        pos.append(w.flag)
    pos=','.join(pos)
    jieba_pos.append(pos)

candidate['jieba_pos']=jieba_pos
display(candidate)
```
