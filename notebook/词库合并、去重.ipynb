```python
import pandas as pd
pre=pd.read_csv(r'E:\毕业设计\词库\百度网络新词.txt', header=None, encoding='utf-8')

# 去重
# pre=pre.drop_duplicates()
# pre.to_csv(r'E:\毕业设计\词库\百度网络新词.txt', header=None,index=None)
pre.columns=['word']
# display(pre)
# 多于四个字的删除
dele=[]
for index,row in pre.iterrows():
    if len(row['word'])>4:
        dele.append(index)
# print(dele)    
# display(pre)
pre=pre.drop(index=dele)
display(pre)
pre.to_csv(r'E:\毕业设计\词库\baidu_new.txt', header=None,index=None)
```
