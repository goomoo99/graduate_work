```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
feature=pd.read_csv(r'../data/room36252/new_flagfeature_final_room36252danmu0318.csv')

feature=feature.iloc[:, 4:9]
display(feature)
data=np.array(feature)
display(data)

```

```python
clf = KMeans(n_clusters=2)
clf.fit(data)  # 分组
centers = clf.cluster_centers_ # 两组数据点的中心点
labels = clf.labels_   # 每个数据点所属分组
print(centers)
print(labels)

x0 = data[labels == 0]
x1 = data[labels == 1]
# x2 = data[label_pred == 2]

plt.figure(figsize=(14, 7))
# 画出聚类结果，三种颜色表示
plt.scatter(x0[:, 2], x0[:, 4], c = "green", marker='o', label='label 0')
plt.scatter(x1[:, 2], x1[:, 4], c = "blue", marker='*', label='label 1')
# plt.scatter(x2[:, 2], x2[:, 3], c = "red", marker='+', label='label 2')
# 画出聚类中心
plt.scatter(centers[:,2],centers[:,4],c='black',marker='D',s=150,alpha=0.75)
plt.xlabel('pos_proportion')
plt.ylabel('neg_proportion')
plt.show()
```

```python
clf = KMeans(n_clusters=3)
clf.fit(data)  # 分组
centers = clf.cluster_centers_ # 两组数据点的中心点
labels = clf.labels_   # 每个数据点所属分组
print(centers)
print(labels)

x0 = data[labels == 0]
x1 = data[labels == 1]
x2 = data[labels == 2]

plt.figure(figsize=(14, 7))
# 画出聚类结果，三种颜色表示
plt.scatter(x0[:, 2], x0[:, 4], c = "green", marker='o', label='label 0')
plt.scatter(x1[:, 2], x1[:, 4], c = "blue", marker='*', label='label 1')
plt.scatter(x2[:, 2], x2[:, 4], c = "red", marker='+', label='label 2')
# 画出聚类中心
plt.scatter(centers[:,2],centers[:,4],c='black',marker='D',s=150,alpha=0.75)
plt.xlabel('pos_proportion')
plt.ylabel('neg_proportion')
plt.show()
```

```python
clf = KMeans(n_clusters=4)
clf.fit(data)  # 分组
centers = clf.cluster_centers_ # 两组数据点的中心点
labels = clf.labels_   # 每个数据点所属分组
print(centers)
print(labels)

x0 = data[labels == 0]
x1 = data[labels == 1]
x2 = data[labels == 2]
x3 = data[labels == 3]

plt.figure(figsize=(14, 7))
# 画出聚类结果，三种颜色表示
plt.scatter(x0[:, 2], x0[:, 4], c = "green", marker='o', label='label 0')
plt.scatter(x1[:, 2], x1[:, 4], c = "blue", marker='*', label='label 1')
plt.scatter(x2[:, 2], x2[:, 4], c = "red", marker='+', label='label 2')
plt.scatter(x3[:, 2], x3[:, 4], c = "yellow", marker='+', label='label 3')
# 画出聚类中心
plt.scatter(centers[:,2],centers[:,4],c='black',marker='D',s=150,alpha=0.75)
plt.xlabel('pos_proportion')
plt.ylabel('neg_proportion')
plt.show()
```
