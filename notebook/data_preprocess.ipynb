```python
import re
import pandas as pd
import numpy as np

df=pd.read_csv(r"../data/room101danmu.csv")
df.columns=['id','time','content']
dfs=df[-20000:]
def find_spec_str(desstr):
    #检测除中英文及数字以外的其他字符
    res = re.compile("[^\u4e00-\u9fa5^a-z^A-Z^0-9]")
    if res.match(desstr)!=None:
        return True
    else:
        return False


# 对于每一行，通过列名name访问对应的元素
for index,row in dfs.iterrows():
    if find_spec_str(row['content'])==True:
        print(index,row['id'], row['content']) # 输出每一行
```
